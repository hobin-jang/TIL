```python
import zipfile
import shutil
from google.colab import files

# 압축할 폴더 이름
folder_name = "llama2-korquad-finetuning-da"

# 생성될 ZIP 파일 이름
zip_file_name = "llama2-korquad-finetuning-da.zip"

# 폴더를 ZIP 파일로 압축
shutil.make_archive(zip_file_name[:-4], 'zip', folder_name)

# ZIP 파일을 로컬로 다운로드
files.download(zip_file_name)
```
