langchain 이용해서 pdf 정보 가져오도록 하기 

```python
from langchain.document_loaders import PyPDFLoader

# PyPDFLoader로 웹에서 pdf 가져오기
loader = PyPDFLoader("https://www.health.kr/Menu.PharmReview/_uploadfiles/%EC%95%84%ED%86%A0%ED%94%BC%ED%94%BC%EB%B6%80%EC%97%BC(Atopic%20Dermatitis)%20(1)_JS.pdf")
# pdf 페이지 단위로 분리
pages = loader.load_and_split()
```

```python
# 첫 페이지 
pages[0]
```

```python
# pdf 내용을 chunk 단위로 분리 
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
splits = text_splitter.split_documents([pages[0]])
```

```python
print(len(splits))
print(type(splits))
print(splits)

# 전체 다 하려니 너무 오래 걸려서 split 된 거 첫 번째만 사용 
splits = splits[0:1]
```

(llama-cpp로 할려니 계속 타입 에러 발생.... 음...!!! embedding 과정에서 뭔가 문제가 있는 듯.)


```python
# chunk된 내용을 embedding 후 chroma db (vector db)에 저장 
from langchain.vectorstores import Chroma
from langchain.embeddings import OllamaEmbeddings

embedding = OllamaEmbeddings(model="llama3:latest")
```

```python
# embedding 되는 지 확인

test_string = "test string !!!"
test_embedded = embedding.embed_query(test_string)

print(test_embedded)
```

```python
# llm model에 맞는 embedding 기법으로 document를 embedding 후 vector db에 저장 
vectorstore = Chroma.from_documents(documents=splits, embedding=embedding)
```

```python
# vector db를 검색 대상으로 사용 
retriever = vectorstore.as_retriever()
```

```python
# 템플릿 지정 
from langchain.prompts import PromptTemplate 

template = """
You are a Korean. So the question is korean, and your response must be korean. 
Furthermore, do not answer if you don't understand or don't know the question, then you always answer 'I don't understand your question' in Korean.
Belows are some information for helping you.
{context}

Question : {question}
"""

rag_prompt = PromptTemplate.from_template(template)
```

```python
# llama 3 적용 
from langchain.chat_models import ChatOllama

llm = ChatOllama(model="llama3:latest")
```

```python
# prompt template 적용한 체인 설정
from langchain.schema.runnable import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
rag_chain = {"context": retriever, "question": RunnablePassthrough()} | rag_prompt | llm | output_parser
```

```python
rag_chain.invoke("아토피피부염의 원인은 무엇이야?")
```
